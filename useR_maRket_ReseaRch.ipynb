```r
# perintah untuk menghapus semua data di environment
# dimulai dari nol
rm(list=ls())

# beberapa libraries yang sudah biasa dipakai untuk data carpentry, machine learning, dan deep learning
# sudah pre installed di sini
# jadi tinggal panggil saja
library(dplyr) # library utk data carpentry
library(tidyr) # library utk data carpentry juga - function pamungkas dari library ini adalah separate()
library(readxl) # library utk import file excel

# sekarang kita akan pakai data contoh bernama mtcars
# = berguna untuk pendefinisian object
data_latihan = mtcars

# kita lihat bagaimana bentuk datanya
data_latihan
```

```r
# pre processing data
# cek struktur data
str(data_latihan)

# kenapa pas kita lihat datanya, ada varian mobilnya?
# tapi pas kita str() tidak ada variabel yang menginformasikan varian mobil
```

```r
# untuk data.frame atau tabel, ada yang namanya colnames() ada yang namanya itu row.names()
# ternyata si varian mobil itu adalah rownames dari data_latihan
row.names(data_latihan)

# sekarang kita akan bikin variabel baru berisi nama varian dari mobil ke data_latihan
data_latihan$varian_mobil = row.names(data_latihan)
```

```r
# sekarang kita bisa hapus row.names - nya
# gak dihapus sebenarnya gak masalah, tapi kalau mau lebih clean datanya, kita hapus aja
row.names(data_latihan) = NULL

# mindahin variabel varian_mobil kedepan, caranya gimana?
# ada fungsi bernama relocate
# caranya kita pake piping %>%
# piping itu artinya then atau maka
data_latihan_baru = 
    data_latihan %>%
    relocate(varian_mobil, .before = mpg)

# cara baca fungsi di atas adalah:
# data_latihan_baru adalah data_latihan YANG saya relocate varian_mobil ke before mpg
```

```r
# kita lihat hasil data yang baru
data_latihan_baru
```

```r
# untuk variabel am sebenarnya dia adalah variabel kategorik
# di R itu ada tipe data yang bernama `factor`
# am = 0 artinya manual
# am = 1 artinya matic
# jadi kita bisa ubah si am ke bentuk factor
data_latihan_baru = 
  data_latihan_baru %>%
  mutate(am = factor(am)
        )

# cara baca skripnya:
# kita akan definisikan data_latihan_baru dari data_latihan_baru yang lama TAPI sudah kita hitung (mutate) 
# yang dimutate adalah am yang tadinya numeric kita ubah jadi factor

# untuk menampilkan statistik dari data_latihan_baru
summary(data_latihan_baru)
```

```r
# kita ingin membuat cross tabulasi
# by am
# berapa sih rata-rata dari mpg?
# berapa standar deviasi dari disp? 
# berapa total dari hp?
# library(expss) ini gunanya utk replace SPSS yang biasa MR pakai
# kita akan pakai library(dplyr) untuk membuat cross tabulasi
tabulasi_1 = 
  data_latihan_baru %>%
  group_by(am) %>%
  summarise(rata_mpg = mean(mpg),
            sdev_disp = sd(disp),
            total_hp = sum(hp)
            ) %>%
  ungroup()

# cara baca skrip di atas:
# saya definisikan tabulasi_1 sebagai:
  # data_latihan_baru YANG
  # saya grup berdasarkan am LALU
  # di-summarise sesuai dengan permintaan
  # ungroup

tabulasi_1
```

```r
# sebelum import data sebaiknya kita cek dulu
# apa yang dicek?
  # ada berapa sheet di sana?
excel_sheets("covid.xlsx") 
```

```r
# import data .xlsx ke dalam environment si R
data_covid = read_excel("covid.xlsx")
```

```r
# langkah standar: selalu cek struktur datanya
str(data_covid)
```

```r
# hal yang harus kita kerjakan
  # 1. kita mau rename nama variabel: Country,Other -> negara
  # 2. saya hanya ingin ambil variabel: negara, TotalCases, ActiveCases
  # 3. saya mau ubah TotalCases dan ActiveCases yang tadinya character ke numeric
  # 4. variabel negara murni hanya nama negara. Gak boleh isinya itu total atau nama benua
  # 5. kita pastikan bahwa tidak ada kolom yang kosong

# Khusus untuk melakukan filter terhadap benua, saya akan bikin variabel baru bernama sensor
sensor = c("Asia","North America","South America","Europe","Africa","Oceania","World","Total:")

# proses pembuatan data yang clean
data_covid_clean = 
  data_covid %>%
  rename(negara = `Country,Other`) %>% # ini proses rename variabel
  select(negara, TotalCases, ActiveCases) %>% # ini proses seleksi variabel
  mutate(TotalCases = gsub("\\,","",TotalCases),
         ActiveCases = gsub("\\,","",ActiveCases),
         TotalCases = as.numeric(TotalCases),
         ActiveCases = as.numeric(ActiveCases)
        ) %>%
  filter(!negara %in% sensor) %>% # filter negara yang diluar sensor
  filter(!is.na(negara)) %>% # filter gak boleh ada yang kosong di negara
  filter(!is.na(TotalCases)) %>% # filter gak boleh ada yang kosong di TotalCases
  filter(!is.na(ActiveCases)) # filter gak boleh ada yang kosong di ActiveCases
```

```r
str(data_covid_clean)
data_covid_clean
```

```r
# sorting data activecases yang tertinggi
# misal kita mau cek bottom 20 negara dengan ActiveCases tertinggi

top_20 = 
  data_covid_clean %>% 
  arrange(desc(ActiveCases)) %>% 
  head(20)

top_20
```

```r
# secara default si R itu gak bisa save ke .xlsx
# kita harus panggil library yang lain untuk bisa save ke .xlsx

# secara default, si R itu bisa mengekspor ke bentuk .csv
write.csv(data_covid_clean,
          "data covid yang sudah dibersihkan.csv")
```

```r
# ternyata kita butuhnya file .xlsx
# kita tinggal install library bernama openxlsx
install.packages("openxlsx")

# kalau sudah terinstall library(openxlsx)
library(openxlsx)
write.xlsx(data_covid_clean,
           "data clean.xlsx")
```

```r
# lanjut donk bikin grafiknya
library(ggplot2)

top_20 %>%
  ggplot(aes(x = reorder(negara,-ActiveCases),
             y = ActiveCases)) +
  geom_col(color = "steelblue",fill = "white") +
  geom_label(aes(label = ActiveCases),
             size = 1.5) +
  theme_minimal() +
  labs(title = "Top 20 Negara dengan Active Cases Tertinggi") +
  theme(axis.title = element_blank(),
        axis.text.y = element_blank(),
        axis.text.x = element_text(angle = 45))
```
